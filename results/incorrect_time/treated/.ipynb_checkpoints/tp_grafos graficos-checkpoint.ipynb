```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
i = 0
f = open('avg_m')
line = f.readline()
mdez3 = []
mdez4 = []
while line:
    if i % 2 == 0:
        mdez3.append(line)
    else:
        mdez4.append(line)
        
    line = f.readline()
    i+=1
    
f.close()
```

```python
i = 0
f = open('avg_r')
line = f.readline()
rdez3 = []
rdez4 = []
while line:
    if i % 2 == 0:
        rdez3.append(line)
    else:
        rdez4.append(line)
        
    line = f.readline()
    i+=1
    
f.close()
```
